```python
df = df.drop(columns=['Churn Reason', 'Churn Category'])
```

```python
cols_internet = [
    'Internet Type', 'Online Security', 'Online Backup',
    'Device Protection Plan', 'Premium Tech Support',
    'Streaming TV', 'Streaming Movies', 'Streaming Music',
    'Unlimited Data'
]

df[cols_internet] = df[cols_internet].fillna('No Internet Service')
```

```python
df['Churn_Flag'] = (df['Customer Status'] == 'Churned').astype(int)
```

```python
df['Tenure_Group'] = pd.cut(
    df['Tenure in Months'],
    bins=[0, 6, 12, 24, 48, 72],
    labels=['0-6', '6-12', '12-24', '24-48', '48-72']
)
df['Tenure_Group']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Churn_Flag', y='Tenure in Months', data=df, color='green')
plt.title("Churn vs Tenure")
plt.show()
```

**Insight:** 
Customers who churn have significantly lower tenure, with churn heavily concentrated in the first 0–6 months, indicating early lifecycle drop-off.

```python
df.groupby('Contract')['Churn_Flag'].mean().plot(kind='bar', color='green')
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

**Insight:** 
Month-to-month customers exhibit the highest churn rates, highlighting lack of long-term commitment as a key driver of churn.

```python
df.groupby('Internet Type')['Churn_Flag'].mean().plot(kind='bar', color='green')
plt.title("Churn Rate by Internet Type")
plt.ylabel("Churn Rate")
plt.show()
```

**Insight:** 
Fiber Optic users show the highest churn (~41%) despite having the highest monthly charges, suggesting a potential mismatch between cost and perceived value.

```python
high_risk = df[
    (df['Internet Type'] == 'Fiber Optic') &
    (df['Contract'] == 'Month-to-Month')
]

high_risk['Churn_Flag'].value_counts().plot(kind='bar', color='green')
plt.title("Churn Distribution in High-Risk Segment")
plt.show()
```

**Insight:**  
Nearly 60% of customers in the high-risk segment (Fiber Optic + Month-to-Month) have churned, indicating severe instability within this group and highlighting it as the most critical target for retention efforts.
